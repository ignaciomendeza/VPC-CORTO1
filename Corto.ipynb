## Task 2

**Inciso 1**

Los métodos tradicionales como SIFT y HOG dependen de descriptores de características diseñados manualmente que encuentran patrones locales de bordes o gradientes. Aunque pueden funcionar en entornos controlados, en un contexto agrícola real pueden ser limitantes, ya que las hojas pueden aparecer con condiciones distintas de iluminación, diferentes ángulos en los que se tomó la foto, o tapadas en parte por otras hojas o ramas. Estas variaciones cambian bastante los gradientes y puntos clave que estos algoritmos usan, lo que hace que haya una pérdida de consistencia en los descriptores y disminuye la capacidad de un clasificador como un SVM para identificar correctamente las enfermedades.

Por otro lado, los modelos de Deep Learning basados en redes convolucionales generan mapas de características jerárquicos que aprenden automáticamente representaciones visuales desde patrones simples hasta estructuras semánticas más complejas. En las primeras capas se encuentran bordes y texturas, mientras que en capas profundas se capturan patrones de alto nivel relacionados con la forma y las manchas que caracterizan una enfermedad. Esta representación jerárquica hace que el modelo sea mucho más completo con las variaciones de iluminación y perspectiva, ya que las características importantes se aprenden directamente de los datos en lugar de que sean definidas manualmente.

**Inciso 2**

Si tuviera que elegir entre AlexNet y VGG-16, elegiría VGG-16 porque su arquitectura es más profunda y modular, y está basada en múltiples convoluciones pequeñas de 3×3. Esta estructura permite aumentar progresivamente el campo receptivo y al mismo tiempo mantiene una organización consistente de bloques convolucionales, lo que hace más fácil el uso de transfer learning para reutilizar características aprendidas en grandes datasets. Además, la presencia de múltiples funciones de activación ReLU introduce más no-linealidad en la red, aumentando su capacidad para aprender representaciones visuales complejas y abstractas. Esto es útil para identificar patrones sutiles de enfermedades en hojas de cultivo.
